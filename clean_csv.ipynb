```python
import pandas as pd 

taiwan_charger = pd.read_csv("data\chargers_taiwan.csv")
taiwan_charger
```

```python
taiwan_charger.info()
```

```python
taiwan_charger_cleaned = taiwan_charger[[
    'name', "longitude", "latitude"
    ]]
taiwan_charger_cleaned
```

```python
taiwan_charger_cleaned['name'].value_counts()
```

```python
taiwan_charger_cleaned.isna().sum()
# Wallahi
```

```python
from sklearn.impute import SimpleImputer

imp_name = SimpleImputer(strategy='constant', fill_value="Unnamed")
taiwan_charger_cleaned_2 = imp_name.fit_transform(taiwan_charger_cleaned)
taiwan_charger_cleaned_2
```

```python
taiwan_charger_cleaned_2 = pd.DataFrame(
    taiwan_charger_cleaned_2, columns=taiwan_charger_cleaned.columns
)
```

```python
taiwan_charger_cleaned_2['name'].value_counts()
```

```python
taiwan_charger_cleaned_2.to_csv('data/taiwan_ev_station_mock.csv', index=False)
```
